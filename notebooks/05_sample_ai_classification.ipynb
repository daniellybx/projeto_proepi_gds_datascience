# Sampling Classified Records

This notebook:
1. Reads the classified input file.
2. Calculates the population size (`N`).
3. Calculates the sample size (`n`) using the proportion sampling formula with finite population correction.
4. Randomly selects `n` rows.
5. Saves the sample to `data/sample_human_in_loop`.


```python
from pathlib import Path
import math

import pandas as pd

# Resolve ROOT independently from the execution directory
ROOT = Path.cwd()
if not (ROOT / 'data').exists():
    ROOT = ROOT.parent

PATH_INPUT = ROOT / 'data' / 'results' / 'agent_outputs' / 'agent_outputs_dataset_sintomas_grupos_classificado.xlsx'
PATH_OUTPUT_DIR = ROOT / 'data' / 'sample_human_in_loop'
PATH_OUTPUT_DIR.mkdir(parents=True, exist_ok=True)
PATH_OUTPUT = PATH_OUTPUT_DIR / 'sample_agent_outputs_datasetxlsx'

PATH_INPUT, PATH_OUTPUT
```

```python
df = pd.read_excel(PATH_INPUT)
N = len(df)
print(f'Population size (N): {N}')
```

```python
# Proportion sampling parameters
z = 1.96   # 95% confidence level
p = 0.5    # maximum variability (conservative scenario)
e = 0.05   # 5% margin of error

# Sample size for infinite population
n0 = (z**2 * p * (1 - p)) / (e**2)

# Finite population correction
n = n0 / (1 + (n0 - 1) / N)
n_sample = min(N, math.ceil(n))

print(f'n0 (without correction): {n0:.2f}')
print(f'Sample size n (with correction): {n_sample}')
```

```python
df_sample = df.sample(n=n_sample, random_state=42)
df_sample.to_excel(PATH_OUTPUT, index=False)

print(f'Sample saved to: {PATH_OUTPUT}')
print(f'Rows in sample: {len(df_sample)}')
df_sample.head()
```
